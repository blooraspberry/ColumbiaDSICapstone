Fuzzy Matching on Names

```python
from difflib import SequenceMatcher
```

```python
col = ['name1', 'name2', 'score']
name_map = pd.DataFrame(columns = col)

for i in names:
    for j in names:
        m = SequenceMatcher(None, i, j)
        if m.ratio() > 0.62 and m.ratio() < 1.0 and i[0:2] == j[0:2]:

            name_map = name_map.append({'name1': i, 'name2': j, 'score': m.ratio()}, ignore_index=True)     
```

```python
name_map.to_csv('name_map.csv')

# after this I'm going to manually inspect the names
```
