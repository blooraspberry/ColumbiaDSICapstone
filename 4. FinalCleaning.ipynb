```python
import pandas as pd
import os
from os import listdir
```

```python
path = '/Users/sharontsao/Desktop/Datasets/EarningsCallAll/'
```

```python
industryFiles = [f for f in listdir(path) if '.csv' in f]
```

```python
for k in range(0, len(industryFiles)):
    
    df = pd.read_csv(path + industryFiles[k], encoding = 'ISO-8859-1')
    df = df[['Company', 'Content', 'Date', 'Industry', 'Section', 'SentimentScore', 'Speaker', 'Type']]

    
    date = df['Date'].unique()
    datelist = []
    for i in range(0, len(date)):
        try:
            if str(date[i]).isnumeric() == True:
                if len(str(date[i])) == 8:
                    
                    datelist.append(str(date[i]))
                    #print(datelist)
        except:
            continue
            
    
    df = df[df['Date'].isin(datelist)]

    df['Date'] = pd.to_datetime(df['Date'], format = '%Y%m%d')

    df['Month'] = df['Date'].dt.month
    df['Year'] = df['Date'].dt.year
    df['Day'] = 1
    
    print(industryFiles[k])
    df.to_csv('/Users/sharontsao/Desktop/Datasets/EarningsCallCleaned/EarningsCall_' + industryFiles[k] + '.csv')
```
